###  Gr√°fico de barras agrupadas para mostrar la distribuci√≥n de las caracter√≠sticas num√©ricas seg√∫n la variable objetivo

```python
binary_features = ["Gender","Polyuria","Polydipsia","sudden weight loss","weakness","Polyphagia","Genital thrush","visual blurring","Itching","Irritability","delayed healing","partial paresis","muscle stiffness","Alopecia","Obesity"]

for feature in binary_features:
    cross_tab = pd.crosstab(diabetes_data[feature], diabetes_data['class'])
    cross_tab.plot(kind='bar')
    plt.title(f'Distribuci√≥n de {feature} seg√∫n la variable objetivo')
    plt.xlabel(feature)
    plt.ylabel('Frecuencia')
    plt.legend(title='class', labels=['No Diabetes', 'Diabetes'])
    plt.show()
```
